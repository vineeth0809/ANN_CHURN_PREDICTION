```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
## data loading
data = pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
data = data.drop(columns=["RowNumber", "CustomerId", "Surname"])
```

```python
data.head()
```

```python
#encoding categorical data
lable_encoder_gender = LabelEncoder()
data["Gender"] = lable_encoder_gender.fit_transform(data["Gender"])
```

```python
data
```

```python
##one hot encoding for geography
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder_geo = OneHotEncoder(sparse_output=False)
```

```python
geo_encoder = one_hot_encoder_geo.fit_transform(data[["Geography"]])
geo_encoder
```

```python
one_hot_encoder_geo.get_feature_names_out(["Geography"])
```

```python
geo_encoder_df = pd.DataFrame(geo_encoder, columns=one_hot_encoder_geo.get_feature_names_out(["Geography"]))
```

```python
data = pd.concat([data, geo_encoder_df], axis=1)
```

```python
data =data.drop(columns=["Geography"])
```

```python
data
```

```python
with open("lable_encoder_gender.pkl", "wb") as f:
    pickle.dump(lable_encoder_gender, f)
with open("one_hot_encoder_geo.pkl", "wb") as f:
    pickle.dump(one_hot_encoder_geo, f)
```

```python
x = data.drop(columns=["Exited"], axis=1)
y = data["Exited"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#scaling the data
scalar = StandardScaler()
x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.transform(x_test)
```

```python
x_train_scaled
```

```python
with open("scalar.pkl", "wb") as f:
    pickle.dump(scalar, f)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
## build ann

model = Sequential(
    [
        Dense(64, activation='relu', input_shape=(x_train.shape[1],)), #hl1 connected with input
        Dense(32, activation='relu'), #hl2
        Dense(1, activation='sigmoid') #output layer
    ]
)
```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
model.compile(optimizer=opt, loss="binary_crossentropy", metrics=["accuracy"])
```

```python
## set tensorboard
logs_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensoboard_callback = TensorBoard(log_dir=logs_dir, histogram_freq=1)
```

```python
## set up early stoping
early_stopping_callback = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
```

```python
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=100, callbacks=[tensoboard_callback, early_stopping_callback])
```

```python
model.save("churn_model.h5")
```
