```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import numpy as np
import pandas as pd
```

```python
model = load_model('churn_model.h5')

with open('lable_encoder_gender.pkl', 'rb') as f:
    label_encoder_gender = pickle.load(f)

with open('one_hot_encoder_geo.pkl', 'rb') as f:
    one_hot_encoder_geo = pickle.load(f)

with open('scalar.pkl', 'rb') as f:
    scaler = pickle.load(f)

```

```python
input_data = {
    'CreditScore': 600,
    'Geography:': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
input_df = pd.DataFrame([input_data])
```

```python
geo_encoder = one_hot_encoder_geo.transform([[input_data['Geography:']]])
geo_ecoder_df = pd.DataFrame(geo_encoder, columns=one_hot_encoder_geo.get_feature_names_out(['Geography']))
geo_ecoder_df
```

```python
input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
```

```python
input_df = pd.concat([input_df.drop('Geography:', axis=1), geo_ecoder_df], axis=1)
```

```python
input_df
```

```python
scaled_input = scaler.transform(input_df)
```

```python
scaled_input
```

```python
probabilities = model.predict(scaled_input)[0][0]
```

```python
if probabilities > 0.5:
    print(f"The customer is likely to churn with a probability of {probabilities:.2f}")
else:
    print(f"The customer is unlikely to churn with a probability of {probabilities:.2f}")
```
